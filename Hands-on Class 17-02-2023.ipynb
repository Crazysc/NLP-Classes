#Create corpus of Natgeo lakshmi shankaran/Raul Dias/Romila Thapar 3 passages, then the next two passages and then take vector.

```python
from sklearn.feature_extraction.text import CountVectorizer as CV
vect= CV(binary=True)
```

```python
corpus=["I have a German Shepherd","German Shepherd is from Germany","Germans love gossipping"]
vect.fit(corpus)
```

```python
vocab= vect.vocabulary_
```

```python
for key in sorted(vocab.keys()):
    print("{}:{}".format(key,vocab[key]))
```

```python
print(vect.transform(["Germany has German Shepherd"]).toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity as CS
```

```python
similar= CS(vect.transform(["Germany has German Shepherd"]).toarray(), vect.transform(["Berlin is lovely"]).toarray())
similar
```
